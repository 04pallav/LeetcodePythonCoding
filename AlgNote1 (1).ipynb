```python
#######################################################################################94. Binary Tree Inorder Traversal
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


#########################################################  ITERATIVE SOLUTION   O(h)
class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if root==None:      # if no root is provided return an empty list
            return []
        else:
            result=[]  
            stack=[]
            node=root       #node variable is initiazed at root
            while node or stack:    #while loop goes on until stack is exhausted. To initialize  we give node
                if node:                 # keep going left until possible
                    stack.append(node)
                    node=node.left
                else:
                    node=stack.pop()    # when we cannot go left any more pop 
                    result.append(node.val)  # store the value after popping
                    node=node.right        # go right whenever left child is None
            return result
```

```python
############################################ QUICK SORT ALGORITHM
1. Choose a pivot last element of array
2. choose divider first element of array
2. iterate over the list and move all elements smaller than pivot to left of pivot,larger to right of pivot
3. if the element is less than pivot swap it with "divider" else move i ahead
4. divider moves forward after each swap 
5. when i reaches end, swap pivot with divider
6. after this step pivot is in correct sorted position
7. right side is all greater than pivot ..left side is all smaller than pivot ...recurse on both sides
8. Note the element which was in pivot will go to its correct position



def partition(array, left, right):
    i = left-1
    for j in range(left, right):
        if array[j] <= array[right]:
            i += 1
            array[j], array[i] = array[i], array[j]
    array[i+1], array[right] = array[right], array[i+1]
    return i+1

def quicksort(array, left, right):
    if left < right:
        pivot = partition(array, left, right)
        quicksort(array, left, pivot-1)
        quicksort(array, pivot+1, right)


array = [3, 2, 1, 0, 9, 8, 7, 6, 5, 4]
quicksort(array, 0, 9)
print array
```

```python
############################################### 54. Spiral Matrix 

class Solution(object):
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        if len(matrix)==0:
            return []      ####edge case
        
        m=len(matrix)
        n=len(matrix[0])
        
        solution=[]
        mini=0                 #### define four variables min max i and j 
                                #### and 4 directions
        maxi=m-1
        minj=0
        maxj=n-1
        i=0
        j=0
        dir='right'
        
        if j==maxj:
            dir='down'     ###############edge case of a single column
        
        while len(solution)<m*n:
            if dir=='right':
                solution.append(matrix[i][j])
                j+=1
                if j==maxj:
                    dir='down'
                    mini+=1
            elif dir=='down':
                solution.append(matrix[i][j])
                i+=1
                if i==maxi:
                    dir='left'
                    maxj-=1
            elif dir=='left':
                solution.append(matrix[i][j])
                j-=1
                if j==minj:
                    dir='up'
                    maxi-=1
            elif dir=='up':
                solution.append(matrix[i][j])
                i-=1
                if i==mini:
                    dir='right'
                    minj+=1
            else:
                print 'never'
            
        return solution
```

```python
#######################################################################################94. Binary Tree Inorder Traversal
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


#########################################################  ITERATIVE SOLUTION   O(h)
class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if root==None:      # if no root is provided return an empty list
            return []
        else:
            result=[]  
            stack=[]
            node=root       #node variable is initiazed at root
            while node or stack:    #while loop goes on until stack is exhausted. To initialize  we give node
                if node:                 # keep going left until possible
                    stack.append(node)
                    node=node.left
                else:
                    node=stack.pop()    # when we cannot go left any more pop 
                    result.append(node.val)  # store the value after popping
                    node=node.right        # go right whenever left child is None
            return result
```

```python
###################################################199. Binary Tree Right Side View 

Do a level order traversal and choose the last node in each level



# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def rightSideView(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        thisLevel=[]
        solution=[]
        if root:
            thisLevel.append(root)
        while len(thisLevel)>0:
            levelValues=[]
            nextLevel=[]
            for node in thisLevel:
                levelValues.append(node.val)
                if node.left:
                    nextLevel.append(node.left)
                if node.right:
                    nextLevel.append(node.right)
        
            solution.append(levelValues[len(levelValues)-1]) ###instead of appending whole list just appended last element
            thisLevel=nextLevel
        return solution
```

```python
############################################ QUICK SORT ALGORITHM
1. Choose a pivot last element of array
2. choose divider first element of array
2. iterate over the list and move all elements smaller than pivot to left of pivot,larger to right of pivot
3. if the element is less than pivot swap it with "divider" else move i ahead
4. divider moves forward after each swap 
5. when i reaches end, swap pivot with divider
6. after this step pivot is in correct sorted position
7. right side is all greater than pivot ..left side is all smaller than pivot ...recurse on both sides
8. Note the element which was in pivot will go to its correct position



def partition(array, left, right):
    i = left-1
    for j in range(left, right):
        if array[j] <= array[right]:
            i += 1
            array[j], array[i] = array[i], array[j]
    array[i+1], array[right] = array[right], array[i+1]
    return i+1

def quicksort(array, left, right):
    if left < right:
        pivot = partition(array, left, right)
        quicksort(array, left, pivot-1)
        quicksort(array, pivot+1, right)


array = [3, 2, 1, 0, 9, 8, 7, 6, 5, 4]
quicksort(array, 0, 9)
print array
```
